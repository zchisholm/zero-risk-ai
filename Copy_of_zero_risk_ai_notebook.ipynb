```python
import pandas as pd
import plotly.graph_objects as go

black_swan_events = [
    {"date": "2000-03-10", "label": "Dotcom Crisis"},
    {"date": "2001-09-11", "label": "9/11"},
    {"date": "2008-09-01", "label": "Financial Crisis"},
    {"date": "2009-11-05", "label": "EU Debt Crisis"},
    {"date": "2011-03-11", "label": "Fukushima Disaster"},
    {"date": "2014-01-01", "label": "Crude Oil Crisis"},
    {"date": "2015-08-24", "label": "China Black Monday"},
    {"date": "2020-03-03", "label": "COVID Pandemic"},
]
# Filter the data where Y == 1
y_equals_1_data = df[df['Y'] == 1]

#'BDIY','VIX',  'LF98TRUU','DXY', 'MXUS'
# Define the columns to plot
columns_to_plot = ['ECSURPUS', 'USGG30YR', 'XAU BGNL', 'Cl1', 'GTDEM10Y', 'USGG2YR', 'MXUS']

# Create a Plotly figure
fig = go.Figure()

# Add line plots for each column
for col in columns_to_plot:
    if col in df.columns:
        fig.add_trace(go.Scatter(
            x=df['Data'],
            y=df[col],
            mode='lines',
            name=f'{col} Stock Price',
            hovertemplate='<b>Date</b>: %{x}<br><b>Price</b>: %{y}<extra></extra>'
        ))

# Add scatter plots for anomalies (Y == 1)
for col in columns_to_plot:
    if col in y_equals_1_data.columns:
        fig.add_trace(go.Scatter(
            x=y_equals_1_data['Data'],
            y=y_equals_1_data[col],
            mode='markers',
            name=f'{col} Anomalies (Y=1)',
            marker=dict(color='blue', size=8),
            hovertemplate='<b>Date</b>: %{x}<br><b>Price</b>: %{y}<extra></extra>'
        ))

# Add vertical lines and annotations for black swan events
for event in black_swan_events:
    event_date = pd.to_datetime(event['date'])
    fig.add_shape(
        type="line",
        x0=event_date,
        x1=event_date,
        y0=df[columns_to_plot].min().min(),  # Minimum Y value
        y1=df[columns_to_plot].max().max(),  # Maximum Y value
        line=dict(color="red", width=2, dash="dash"),
    )
    fig.add_annotation(
        x=event_date,
        y=df[columns_to_plot].max().max(),
        text=event['label'],
        showarrow=False,
        yshift=10,
        font=dict(color="red", size=10),
    )

# Customize the layout
fig.update_layout(
    title="Stock Prices with Anomalies Highlighted (Y = 1)",
    xaxis_title="Date",
    yaxis_title="Stock Price",
    legend_title="Legend",
    template="plotly_white",
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    hovermode="x unified"
)

# Show the plot
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
# List of black swan event dates and descriptions
black_swan_events = [
    #{"date": "2000-03-10", "label": "Dotcom Crisis"},
    #{"date": "2001-09-11", "label": "9/11"},
    #{"date": "2008-09-01", "label": "Financial Crisis"},
    #{"date": "2009-11-05", "label": "EU Debt Crisis"},
    # {"date": "2011-03-11", "label": "Fukushima Disaster"},
    # {"date": "2014-01-01", "label": "Crude Oil Crisis"},
    # {"date": "2015-08-24", "label": "China Black Monday"},
     {"date": "2020-03-03", "label": "COVID Pandemic"},
]

df['Data'] = pd.to_datetime(df['Data'])

# Define the date range for filtering
start_date = "2019-12-01"
end_date = "2021-12-31"

# Filter the DataFrame for the date range
filtered_df = df[(df['Data'] >= start_date) & (df['Data'] <= end_date)]

# Filter the data where Y == 1
y_equals_1_data = filtered_df[filtered_df['Y'] == 1]
#'ECSURPUS', 'USGG30YR',  'GTDEM10Y', 'USGG2YR'
#'BDIY','VIX',  'LF98TRUU','DXY', 'MXUS', 'XAU BGNL', 'Cl1',
# Define the columns to plot
columns_to_plot = [ 'EONIA','GTITL10YR', 'GTJPY10YR', 'MXUS' ]

# Create a Plotly figure
fig = go.Figure()

# Add line plots for each column
for col in columns_to_plot:
    if col in df.columns:
        fig.add_trace(go.Scatter(
            x=filtered_df['Data'],
            y=filtered_df[col],
            mode='lines',
            name=f'{col} Stock Price',
            hovertemplate='<b>Date</b>: %{x}<br><b>Price</b>: %{y}<extra></extra>'
        ))

# Add scatter plots for anomalies (Y == 1)
for col in columns_to_plot:
    if col in y_equals_1_data.columns:
        fig.add_trace(go.Scatter(
            x=y_equals_1_data['Data'],
            y=y_equals_1_data[col],
            mode='markers',
            name=f'{col} Anomalies (Y=1)',
            marker=dict(color='blue', size=8),
            hovertemplate='<b>Date</b>: %{x}<br><b>Price</b>: %{y}<extra></extra>'
        ))

# Add vertical lines and annotations for black swan events
for event in black_swan_events:
    event_date = pd.to_datetime(event['date'])
    fig.add_shape(
        type="line",
        x0=event_date,
        x1=event_date,
        y0=filtered_df[columns_to_plot].min().min(),  # Minimum Y value
        y1=filtered_df[columns_to_plot].max().max(),  # Maximum Y value
        line=dict(color="red", width=2, dash="dash"),
    )
    fig.add_annotation(
        x=event_date,
        y=filtered_df[columns_to_plot].max().max(),
        text=event['label'],
        showarrow=False,
        yshift=10,
        font=dict(color="red", size=10),
    )


# Customize the layout
fig.update_layout(
    title="Stock Prices with Anomalies Highlighted (Y = 1)",
    xaxis_title="Date",
    yaxis_title="Stock Price",
    legend_title="Legend",
    template="plotly_white",
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    hovermode="x unified"
)



# Show the plot
fig.show()
```
